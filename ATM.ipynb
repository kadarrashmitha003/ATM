```python
val_pin = 4321

def deposit(balance):
    amount_deposit = int(input("Enter your amount to deposit: "))
    if amount_deposit % 100 == 0:
        if amount_deposit <= 2000:
            balance += amount_deposit
            print("Amount is deposited successfully.")
            balance_enquiry(balance)
        else:
            print("Deposit limit exceeded (Max ₹2000 allowed).")
    else:
        print("Invalid deposit! Enter in multiples of ₹100.")
    return balance

def withdraw(balance):
    amount_withdraw = int(input("Enter your amount to withdraw: "))
    if amount_withdraw % 100 == 0:
        if amount_withdraw <= balance:
            if amount_withdraw <= 20000:
                balance -= amount_withdraw
                print(amount_withdraw, "withdrawn successfully.")
                balance_enquiry(balance)
            else:
                print("Withdrawal limit exceeded! (Max ₹20000 allowed)")
        else:
            print("Insufficient balance.")
    else:
        print("Withdrawal must be in multiples of ₹100.")
    return balance

def balance_enquiry(balance):
    print("Available Balance: ₹", balance)

def validation():
    pin = int(input("Enter the PIN: "))
    if pin == val_pin:
        balance = 500000
        while True:
            print("\nChoose an option:")
            print("1) Deposit")
            print("2) Withdraw")
            print("3) Balance Enquiry")
            print("4) Exit")
            
            choice = int(input("Enter your choice (1-4): "))
            
            if choice == 1:
                balance = deposit(balance)
            elif choice == 2:
                balance = withdraw(balance)
            elif choice == 3:
                balance_enquiry(balance)
            elif choice == 4:
                print("Thank you for using the ATM. Goodbye!")
                break
            else:
                print("Invalid choice.")
    else:
        print("Invalid PIN.")

validation()
```
